```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../lightning_logs/version_23_resnet18_binary/metrics.csv')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train_loss_epoch'], label='Training Loss')
plt.plot(df['epoch'], df['val_loss_epoch'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
#plt.yscale('log')
plt.title('Training and Validation Loss')
plt.legend()
plt.savefig('loss.png')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train_acc_epoch'], label='Training Accuracy')
plt.plot(df['epoch'], df['val_acc_epoch'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.savefig('accuracy.png')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['learning_rate'], label='Learning Rate')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate')
plt.legend()
plt.show()
```
